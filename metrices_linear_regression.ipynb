```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/placement.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package (in lakhs)')
plt.show()
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_train, lr.predict(x_train), color='black')
plt.xlabel('CGPA')
plt.ylabel('Package (in lakhs)')
plt.show()
```

```python
# calculating different metrices

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

#here actual value and predicted values

y_predict = lr.predict(x_test)
y_actual = y_test
```

```python
# calculating MAE
print("MAE",mean_absolute_error(y_actual,y_predict))

# calculating MSE
print("MSE",mean_squared_error(y_actual,y_predict))

#calculatinr RMSE
print("RMSE",np.sqrt(mean_squared_error(y_actual,y_predict)))

#calculating r2_score
print("R2_score",r2_score(y_actual,y_predict))

r2 = r2_score(y_actual,y_predict)

print(x_test.shape)

#calculating abosolute r2_score
print("Absolute R2 Score", 1-((1-r2)*(40-1)/(40-1-1)))
```
